```python
# Problem Statement 1: 
 
# Is gender independent of education level? A random sample of 395 people were surveyed and each person was asked to report the highest education level they obtained. The data that resulted from the survey is summarized in the following table: 
# High School    Bachelors   Masters   Ph.d.   Total 
# Female 60             54          46        41      201 
# Male   40             44          53        57      194 
# Total  100            98          99        98      395 
 
# Question:  Are gender and education level dependent at 5% level of significance?  
# In other words, given the data collected above, is there a relationship between the gender of an individual 
# and the level of education that they have obtained? 

import scipy.stats as st

print("Let, Null Hypothesis H0: Gender and education level independent at 5% level of significance")
print("Let, Alternate Hypothesis Ha: Gender and education level dependent at 5% level of significance")
#From given table
female = [60, 54, 46, 41] 
male = [40, 44, 53, 57] 
female_total = 201 
male_total = 194 
people_total = 395 
edu_level_total = [100, 98, 99, 98] 

#Given, 5% level of significance
significance_level = 0.05 

#Given, Education levels: High School, Bachelors, Masters & Ph.d
sample_size = 4 

def expected_values_calculation(comm_total_lst, seq_total, sample_len): #Calculates expected values 
    lst = []
    for val in comm_total_lst:
        lst.append(val*seq_total/sample_len)
    return lst

#Calculates expected values for female
expected_female = expected_values_calculation(edu_level_total,female_total,people_total) 

#Calculates expected values for male
expected_male = expected_values_calculation(edu_level_total,male_total,people_total) 

f_obs = female + male 
print("F-Observed (From given table):",f_obs)

f_exp = expected_female + expected_male
print("F-Expected (Calculated from given table):",f_exp)

dof = sample_size - 1 #Degree of freedom of given data set

chi_sq_calc = st.chisquare(f_obs,f_exp) #Calculated Chi-Square for given observed and expected values 
print("Chi-Square (Calculated using F-Observed and F-Expected):",chi_sq_calc.statistic)

chi_sq_critical = st.chi2.isf(q=significance_level, df=dof) #Calculates Chi-Square critical values 
print("Chi-Square Critical (Calculated using given significance level and degree of freedom):",chi_sq_critical)

#The critical value of Chi-Square with calculated degree of freedom is lesser than the calculated.Therefore we reject the null hypothesis and conclude that the education level depends on gender at a 5% level of significance.
print("If the critical value of Chi-Square with calculated degree of freedom is lesser than the calculated. Therefore we reject the null hypothesis and conclude that the education level depends on gender at a 5% level of significance.")

print("==========================================Solution==========================================")
if chi_sq_critical < chi_sq_calc.statistic:
    print("Reject the null hypothesis and this concludes that the education level depends on gender at a 5% level of significance.")
else:
    print("Null hypothesis can not be rejected and this concludes that the education level does not depends on gender at a 5% level of significance.")
```

```python
# Problem Statement 2: 
 
# Using the following data, perform a oneway analysis of variance using α=.05. Write up the results in APA format. 
 
# [Group1: 51, 45, 33, 45, 67] [Group2: 23, 43, 23, 43, 45] [Group3: 56, 76, 74, 87, 56] 

import scipy.stats as st
import statistics as sts

print("Let, Null Hypothesis H0: μ1 = μ2 = μ3 ")
print("   ")
print("Let, Alternate Hypothesis Ha: μ1 ≠ μ2 ≠ μ3 ") 
print("   ")
print("[Group1: 51, 45, 33, 45, 67]")
print("[Group2: 23, 43, 23, 43, 45]")
print("[Group3: 56, 76, 74, 87, 56]")
print("   ")
    
group_1 = [51, 45, 33, 45, 67] #Given, [Group1: 51, 45, 33, 45, 67]
group_2 = [23, 43, 23, 43, 45] #Given, [Group2: 23, 43, 23, 43, 45]
group_3 = [56, 76, 74, 87, 56] #Given, [Group3: 56, 76, 74, 87, 56]

n = len(group_1) + len(group_2) + len(group_3) #Count of all the availabale data 
k = 3 #Given three group of samples
significance_level = 0.05 #Given, using α=.05

group_1_mean = sts.mean(group_1) #Mean value of Group1
group_2_mean = sts.mean(group_2) #Mean value of Group2
group_3_mean = sts.mean(group_3) #Mean value of Group3
total_mean = sts.mean(group_1 + group_2 + group_3) ##Mean value of all the given groups

dof_between = k-1 #Degree of freedom between
print("Degree of freedom between (dof_between):",dof_between)
print("   ")

dof_within= n-k #Degree of freedom within
print("Degree of freedom within (dof_within):",dof_within)
print("   ")

group_size = len(group_1) #From given group

#Sum of Squares Between
ss_between = (((group_1_mean-total_mean)**2) + ((group_2_mean-total_mean)**2) + ((group_3_mean-total_mean)**2)) * group_size
print("Sum of Squares Between (ss_between):",ss_between)
print("   ")

ms_between = ss_between/dof_between #Mean Square Between
print("Mean Square Between (ms_between = ss_between/dof_between):",ms_between)
print("   ")

#Calculates sum of squares for given sequence and mean.
def calc_sum_of_square(sequence, mean): 
    sum_val = 0
    for val in sequence:
        sum_val += ((val-mean)**2)
    return sum_val

#Sum of Squares Within
ss_within = calc_sum_of_square(group_1, group_1_mean) + calc_sum_of_square(group_2, group_2_mean) + calc_sum_of_square(group_3, group_3_mean)
print("Sum of Squares Within (ss_within):",ss_within)
print("   ")

ms_within = ss_within/dof_within #Mean Square Within
print("Mean Square Within (ms_within = ss_within/dof_within):",ms_within) 
print("   ")

F_Calc = ms_between/ms_within #F-Satistic
print("F-Calculated (ms_between/ms_within):",F_Calc)
print("   ")

#Calculating F-Critical using given significance level and degree of freedom
F_Critical = st.f.ppf(q=1-significance_level, dfn=dof_between, dfd=dof_within) 
print("F-Critical:",F_Critical)
print("   ")

print("If the critical value of F from tables is less than the calculated value of F, reject the null hypothesis")
print("--------------------------------------------------Result-------------------------------------------------")
print("   ")
if F_Critical < F_Calc:
    print("Reject Null Hypothesis H0 and conclude that at least two of the means are significantly different from each other.")
else:
    print("Null Hypothesis H0 can not be rejected and conclude that all the means are equal.")
print("   ")
```

```python
# Problem Statement 3: 
 
# Calculate F Test for given 10, 20, 30, 40, 50 and 5,10,15, 20, 25.  
 
# For 10, 20, 30, 40, 50: 
 
import scipy.stats as st
import statistics as sts
#Null Hypothesis, Ho: No difference in variances.
print("Let, Null Hypothesis H0: No difference in variances.") 
print("   ")
#Alternate Hypothesis, Ha: Difference in variances.
print("Let, Alternate Hypothesis Ha: Difference in variances.\n") 
print("   ")

#Given data
lst_1 = [ 10, 20, 30, 40, 50] 
print("Given first data set:",lst_1)
print("   ")
lst_2 = [ 5, 10, 15, 20, 25] 
print("Given second data set:",lst_2)
print("   ")
#Variance of 10, 20, 30, 40, 50
var_1 = sts.variance(lst_1) 
print("Variance of 10, 20, 30, 40, 50 is (var_1):",var_1)
print("   ")
#Variance of 5, 10, 15, 20, 25
var_2 = sts.variance(lst_2) 
print("Variance of 5, 10, 15, 20, 25 is (var_2):", var_2)
print("   ")
#F Test = (variance of 10, 20, 30, 40, 50) / (variance of 5, 10, 15, 20, 25)
F_Test = var_1/var_2  
print("F-Test value of the given data sets (var_1/var_2):",F_Test)
print("   ")
#Degrees of freedom 10, 20, 30, 40, 50
dof_lst_1 = len(lst_1)-1 
print("Degree of freedom of 10, 20, 30, 40, 50 is:",dof_lst_1)
print("   ")
#Degrees of freedom 5, 10, 15, 20, 25
dof_lst_2 = len(lst_2)-1 
print("Degree of freedom of 5, 10, 15, 20, 25 is:",dof_lst_2)
print("   ")
significance_level = 0.05 #If not given, default alpha value = 0.05

# F-Table value using default alpha value = 0.05 and calculated degree of freedom
F_Critical = st.f.ppf(q=1-significance_level, dfn=dof_lst_1, dfd=dof_lst_2) 
print("F-Critical value using default alpha (value = 0.05) is: "+str(F_Critical))
print("   ")
#If the F-Table value is smaller than the calculated value, you can reject the null hypothesis.
print("If the F-Crtitical value is smaller than the calculated value, we can reject the null hypothesis.")
print("   ")
print("------------------------Result-----------------------")
print("   ")
if F_Critical < F_Test:
    print("Null hypothesis can be rejected.")
else:
    print("Null hypothesis can not be rejected.")
print("   ")
```
